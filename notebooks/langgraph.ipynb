```python
from typing import TypedDict
from langgraph.graph import StateGraph, START, END
```

```python
# Definir el estado
class State(TypedDict):
    numero1: int
    numero2: int
    suma: int   
    resultado: str

graph = StateGraph(State)
```

```python
# Definir los nodos del workflow
def caso_suma(state):
    suma = state['numero1'] + state['numero2']
    return {'suma': suma}

def caso_par(state):
    return {f'resultado': f'El numero sumado {state["suma"]} es par'}

def caso_impar(state):
    return {'resultado': f'El numero sumado {state["suma"]} es impar'}
```

```python

graph.add_node("Suma", caso_suma)
graph.add_node("Par", caso_par)
graph.add_node("Impar", caso_impar)
```

```python
# Definir la funcion de routing para decidir la rama de ejecucion
def decidir_rama(state):
    if state["suma"] % 2 == 0:
        return "Par"
    else:
        return "Impar"
```

```python
graph.add_edge(START, "Suma")
graph.add_conditional_edges(
    "Suma", 
    decidir_rama,
    {
        "Par": "Par",
        "Impar": "Impar"
    }
)
graph.add_edge("Par", END)
graph.add_edge("Impar", END)
```

```python
compiled = graph.compile()
```

```python
# Probar el grafo con ejemplos
print(compiled.invoke({"numero1": 5, "numero2": 6})["resultado"])
```
